```python
import random
number_of_potions = 3
my_health_points = health_points_of_the_enemy = 50
separator = "_"*25 
while my_health_points > 0 and health_points_of_the_enemy > 0:
    my_input = input("Souhaitez vous attaquer (1) ou utiliser une potion (2) ? ")
    if my_input == "1":
        hits_of_the_enemy = random.randint(5,15)
        my_health_points -= hits_of_the_enemy 
        my_hits = random.randint(5,10)
        health_points_of_the_enemy -= my_hits
        print(f"Vous avez infligé {my_hits} points de dégats à l'ennemi.")
        print(f"""L'ennemi vous a infligé {hits_of_the_enemy} points de dégats.
Il vous reste {my_health_points} points de vie.
Il reste {health_points_of_the_enemy} points de vie à l'ennemi.
{separator}""")
    elif my_input == "2":
        if number_of_potions > 0:
            hits_of_the_enemy = random.randint(5,15)
            my_health_points -= hits_of_the_enemy 
            my_extra_health_points = random.randint(15,50)
            my_health_points += my_extra_health_points
            number_of_potions -= 1
            print(f"Vous récupérez {my_extra_health_points} points de vie ({number_of_potions} potion(s) restantes).")
            print(f"""L'ennemi vous a infligé {hits_of_the_enemy} points de dégats.
Il vous reste {my_health_points} points de vie.
Il reste {health_points_of_the_enemy} points de vie à l'ennemi.
{separator}""")
            hits_of_the_enemy = random.randint(5,15)
            my_health_points -= hits_of_the_enemy
            print("Vous passez votre tour...")
            print(f"""L'ennemi vous a infligé {hits_of_the_enemy} points de dégats.
Il vous reste {my_health_points} points de vie.
Il reste {health_points_of_the_enemy} points de vie à l'ennemi.
{separator}""")
        else:
            print("Vous n'avez plus de potions...", end = "\n")     
    else:
        print("Veuillez entrer une option valide.", end ="\n")
else:
    if my_health_points >= 0:
        print("Tu as gagné")
    else: 
        print("Tu as perdu")
    print("Fin du jeu.")
        
            
        
            
```
